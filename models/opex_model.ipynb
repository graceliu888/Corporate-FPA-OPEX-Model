```python
import pandas as pd

def calculate_variance(df):
    df["Budget_vs_Actual"] = df["Actuals"] - df["Budget"]
    df["Forecast_vs_Actual"] = df["Actuals"] - df["Forecast"]
    df["Budget_vs_Forecast"] = df["Forecast"] - df["Budget"]
    return df

def variance_summary(df):
    return df.groupby("CostCenter")[
        ["Budget_vs_Actual", "Forecast_vs_Actual", "Budget_vs_Forecast"]
    ].sum().reset_index()
```
